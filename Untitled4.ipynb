```python
import numpy as np

# Yields (interest rates) are measured in percentage points
# Scenario: Investors are panicking about the short term, driving short-term yields up
yield_2_year = 4.85   # 2-Year Bond yield at 4.85%
yield_10_year = 4.21  # 10-Year Bond yield at 4.21%

print("Macroeconomic Yield parameters initialized.")
print(f"2-Year Sovereign Yield: {yield_2_year}% | 10-Year Sovereign Yield: {yield_10_year}%")
```

```python
# The mathematical spread calculation
yield_spread = yield_10_year - yield_2_year

print("--- BOND MARKET spread ANALYSIS ---")
print(f"Calculated 10-2 Yield Spread: {yield_spread:.2f}%")
```
